```sql
CREATE TABLE if NOT EXISTS fact_bookings (
  dim_bookingskey INT,
  dim_airportskey INT,
  dim_passengerskey INT,
  dim_flightskey INT,
  amount DECIMAL(10,2),
  booking_date DATE,
  PRIMARY KEY (dim_bookingskey)
);
INSERT INTO fact_bookings (
  dim_bookingskey, dim_airportskey, dim_passengerskey, dim_flightskey, amount, booking_date
)
SELECT 
  b.dim_bookingskey,
  a.dim_airportskey,
  p.dim_passengerskey,
  f.dim_flightskey,
  b.amount,
  b.booking_date
FROM dim_bookings b
LEFT JOIN dim_airports a ON b.airport_id = a.airport_id
LEFT JOIN dim_passengers p ON b.passenger_id = p.passenger_id
LEFT JOIN dim_flights f ON b.flight_id = f.flight_id
WHERE b.updated_date > (SELECT coalesce(max(booking_date),'2025-09-29T06:04:33.908+00:00') FROM fact_bookings);

```
