```python
dbutils.widgets.text('folder_name','')
```

```python
folder_name=dbutils.widgets.get('folder_name')
```

```python
print(folder_name)
```

```python
df = spark.readStream.format('cloudFiles') \
    .option('cloudFiles.format', 'csv') \
    .option('cloudFiles.schemaLocation', f'/Volumes/project/bronze/bronze_files/{folder_name}/checkpoint') \
    .option('cloudFiles.schemaEvolutionMode', 'rescue') \
    .load(f'/Volumes/project/default/raw/{folder_name}/')
```

```python
df.writeStream.format('delta') \
    .outputMode('append') \
    .option('checkpointLocation', f'/Volumes/project/bronze/bronze_files/{folder_name}/checkpoint') \
    .trigger(once=True) \
    .option('path', f'/Volumes/project/bronze/bronze_files/{folder_name}/data/') \
    .start()
```
